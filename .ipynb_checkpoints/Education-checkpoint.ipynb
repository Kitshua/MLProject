```python
education_copy = pd.DataFrame.copy(education)
```

```python
education_copy["Latitude"] = 0
education_copy["Longitude"] = 0
```

```python
for k in range(0, len(education_copy)):
    
    if(education_copy.loc[k, "Province"] == "Punjab"):
        
        education_copy.loc[k, "Latitude"] = 31.1704
        education_copy.loc[k, "Longitude"] = 72.7097
        
    elif(education_copy.loc[k, "Province"] == "Balochistan"):
    
        education_copy.loc[k, "Latitude"] = 28.4907
        education_copy.loc[k, "Longitude"] = 65.0958
        
    elif(education_copy.loc[k, "Province"] == "KP"):

        education_copy.loc[k, "Latitude"] = 34.9526
        education_copy.loc[k, "Longitude"] = 72.3311

    elif(education_copy.loc[k, "Province"] == "Sindh"):

        education_copy.loc[k, "Latitude"] = 25.8943
        education_copy.loc[k, "Longitude"] = 68.5247

    elif(education_copy.loc[k, "Province"] == "FATA"):

        education_copy.loc[k, "Latitude"] = 33.000
        education_copy.loc[k, "Longitude"] = 70.16667
        
    elif(education_copy.loc[k, "Province"] == "AJK"):

        education_copy.loc[k, "Latitude"] = 33.8433
        education_copy.loc[k, "Longitude"] = 73.8513
        
    elif(education_copy.loc[k, "Province"] == "GB"):

        education_copy.loc[k, "Latitude"] = 35.35
        education_copy.loc[k, "Longitude"] = 75.9
        
    else:

        education_copy.loc[k, "Latitude"] = 33.6833
        education_copy.loc[k, "Longitude"] = 73.0833
        
        
    
#         year, month, day = copy_pak_table["date"][k].split('-')
#     copy_pak_table.loc[k, "year"] = year
#     copy_pak_table.loc[k, "month"] = month
#     copy_pak_table.loc[k, "day"] = day
```
